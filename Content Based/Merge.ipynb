```python
pitch['artist'] = pitch['artist'].str.replace(' ', '').str.lower()

# Displaying the updated DataFrame
pitch.head()
```

```python
columns_to_remove = ['album', 'date', 'release_year', 'label', 'link', 'bnm', 'role']
pitch.drop(columns=columns_to_remove, inplace=True)
pitch.head()
```

```python
pitch['review'] = pitch['review'].astype(str)
```

```python
# Separately aggregating 'score' and 'review', then merging the results
score_agg = pitch.groupby(['artist', 'author'])['score'].mean().reset_index()
review_agg = pitch.groupby(['artist', 'author'])['review'].apply(' '.join).reset_index()
genre_agg = pitch.groupby(['artist', 'author'])['genre'].first().reset_index()
```

```python
# Merging the aggregated data
grouped_df = pd.merge(score_agg, review_agg, on=['artist', 'author'])
grouped_df = pd.merge(grouped_df, genre_agg, on=['artist', 'author'])
# Renaming the 'score' column to 'average_score'
grouped_df.rename(columns={'score': 'average_score'}, inplace=True)
```

```python
spotify_df = pd.read_csv("spotify.csv")
spotify_df.columns
```

```python
spotify_df = spotify_df.rename(columns={'artists': 'artist'})
spotify_df.columns
spotify_df.info
```

```python
spotify_df.info()
```

```python
# Remove duplicate rows based on 'artist' and 'Track Name' columns
spotify_df_unique = spotify_df.drop_duplicates(subset=['artist'])
```

```python
def process_artist_name(artist_list_str):
    # Convert the string representation of list to actual list
    artist_list = eval(artist_list_str)
    # Take the first artist, make it lowercase and remove spaces
    return artist_list[0].lower().replace(' ', '')
```

```python
spotify_df_unique['processed_artist'] = spotify_df_unique['artist'].apply(process_artist_name)
```

```python
spotify_df_unique.head()
```

```python
spotify_df_unique = spotify_df_unique.drop(columns=['artist','year','explicit','id','mode','release_date','name'])
spotify_df_unique.head()
```

```python
spotify_df_unique.rename(columns={'processed_artist': 'artist'}, inplace=True)
spotify_df_unique
```

```python
# Grouping by artist and calculating the mean for the specified features
features_to_average = ['valence','acousticness', 'danceability', 'energy', 'instrumentalness', 'key', 'liveness',
                       'loudness', 'popularity', 'speechiness']
artist_grouped_df = spotify_df_unique.groupby('artist')[features_to_average].mean().reset_index()
artist_grouped_df.head()
```

```python
artist_grouped_df['artist'].nunique()
```

```python
artist_grouped_df.to_csv('artist_grouped_df.csv', index=False)
```
