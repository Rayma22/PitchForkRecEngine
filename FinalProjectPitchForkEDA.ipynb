# **Visualizations **

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Set the aesthetic style of the plots
sns.set_style("whitegrid")

# Plot the distribution of album scores
plt.figure(figsize=(10, 6))
sns.histplot(df_new['score'], bins=20, kde=False)
plt.title('Distribution of Album Scores')
plt.xlabel('Score')
plt.ylabel('Count')
plt.show()
```

The album scores appear to be roughly normally distributed with a slight left skew, indicating that there are more albums with higher scores than lower ones. Most albums have scores between 6 and 8, with a peak around the 7 to 7.5 score range.

```python
plt.figure(figsize=(14, 7))
sns.countplot(data=df_new, x='date', color='skyblue')
plt.title('Number of Reviews Published Each Year')
plt.xlabel('Year')
plt.ylabel('Number of Reviews')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

There is a noticeable trend where the number of reviews increases over the years, peaking around the mid-2000s, and then there's a sharp decline after 2011.

```python
# Pie chart for the frequency of "Best New Music"
best_new_music_counts = df_new['bnm'].value_counts()

plt.figure(figsize=(8, 8))
plt.pie(best_new_music_counts, labels=['Not Best New Music', 'Best New Music'], autopct='%1.1f%%', startangle=140)
plt.title('Proportion of Albums Marked as "Best New Music"')
plt.show()
```

only a small fraction of albums (approximately 5%) are designated as "Best New Music" compared to those that are not.

```python
plt.figure(figsize=(14, 7))
genre_counts = df_new['genre'].value_counts()
sns.barplot(x=genre_counts.index, y=genre_counts.values, palette='viridis')
plt.title('Number of Reviews by Genre')
plt.xlabel('Genre')
plt.ylabel('Number of Reviews')
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```

This shows the number of reviews by genre -> It appears that genres like "rock," "electronic," and "experimental" have the highest number of reviews, while other genres have considerably fewer reviews.

 Checking the relationship between the score and the "Best New Music" designation to see if higher scores are associated with this designation

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x='bnm', y='score', data=df_new)
plt.title('Score Distribution by "Best New Music" Designation')
plt.xlabel('Best New Music')
plt.ylabel('Score')
plt.xticks([0, 1], ['No', 'Yes'])
plt.show()
```

 There's a clear difference between the two groups: albums that are designated as "Best New Music" tend to have higher scores, with a higher median and a narrower interquartile range. This indicates that not only do "Best New Music" albums score higher on average, but their scores are also more consistently high.
